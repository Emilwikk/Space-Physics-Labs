**Velocity as theoretically predicted**

The theoretically predicted velocity is different to the measured velocity. This is due to the fact than that a prediction is never perfect. Also, the theoretically predicted velocity is only due to drift velocity, while the measured data has more velocity components. Such as gyration and movement along the B-field lines.

```python
# Interpolate E so that the data from B and E is "taken" at the same moment in time
import pandas as pd

# Build DataFrames with datetime index
dfE = pd.DataFrame(E_vec, index=pd.to_datetime(timeTagsE))
dfB = pd.DataFrame(B_vec, index=pd.to_datetime(timeTagsB))

# Convert to SI-units
dfE = dfE*10**-3
dfB = dfB*10**-9

B_SI = dfB.to_numpy()

# Interpolate E onto B timestamps
E_interp = dfE.reindex(dfB.index).interpolate(method="time").values

ExB = np.cross(E_interp,B_SI)
B2 = np.sum(B_SI**2, axis=1, keepdims=True)
predVel = ExB/B2

# Convert to km/s
predVel = predVel*10**-3

# Visualize
plt.figure()
plt.title("Theoretically predicted plasma velocity due to ExB drift.")
plt.xlabel("Time (date)")
plt.ylabel("Velocity (km/s)")
plt.plot(timeTagsB,predVel)
plt.xticks(rotation=70)
plt.show()

plt.figure()
plt.xlabel("Time (date)")
plt.ylabel("Velocity (km/s)")
plt.title("Measured plasma velocity")
plt.plot(timeTagsPart,plasmaVel)
plt.xticks(rotation=70)
plt.show()
```
